```python
# ================================================================
# Import Libraries
# ================================================================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# ================================================================
# Load / Create BMW Sales Dataset
# ================================================================
data = {
    "Year": [2018, 2018, 2019, 2019, 2020, 2020, 2021, 2021, 2022, 2022, 2023, 2023],
    "Model": ["BMW 3 Series", "BMW X5", "BMW 3 Series", "BMW X5",
              "BMW 3 Series", "BMW X5", "BMW 3 Series", "BMW X5",
              "BMW 3 Series", "BMW X5", "BMW 3 Series", "BMW X5"],
    "Region": ["Europe", "Europe", "Europe", "Europe",
               "North America", "North America", "North America", "North America",
               "Asia", "Asia", "Asia", "Asia"],
    "Units_Sold": [85000, 67000, 92000, 70000, 88000, 73000, 96000, 77000, 102000, 83000, 108000, 87000],
    "Revenue_Million_USD": [3400, 4700, 3800, 4900, 3700, 5200, 4000, 5500, 4300, 6000, 4500, 6400]
}

df = pd.DataFrame(data)

# Preview
df.head()
```

```python
# ================================================================
# Basic Data Analysis with Pandas
# ================================================================

# Summary statistics
df.describe()

# Count of unique models and regions
print("Unique Models:", df['Model'].nunique())
print("Unique Regions:", df['Region'].nunique())

# Total sales per region
df.groupby('Region')['Units_Sold'].sum()
```

```python
# ================================================================
# Matplotlib Visualizations
# ================================================================

# Bar Chart â€“ Total Sales by Region
region_sales = df.groupby('Region')['Units_Sold'].sum().sort_values()
region_sales.plot(kind='bar', color=['skyblue', 'lightgreen', 'salmon'])
plt.title("Bar Chart - BMW Sales by Region")
plt.ylabel("Units Sold")
plt.show()

'''Shows which regions contribute most to BMW car sales.
Europe and North America are key markets, but Asia is growing strongly.'''
```

```python
# Line Chart â€“ Sales Trend Over Years (3 Series vs X5)
plt.plot(df[df['Model'] == "BMW 3 Series"]['Year'],
         df[df['Model'] == "BMW 3 Series"]['Units_Sold'],
         marker='o', label='BMW 3 Series', color='royalblue')

plt.plot(df[df['Model'] == "BMW X5"]['Year'],
         df[df['Model'] == "BMW X5"]['Units_Sold'],
         marker='s', label='BMW X5', color='orange')

plt.title("Line Chart - Sales Trend (2018â€“2023)")
plt.xlabel("Year")
plt.ylabel("Units Sold")
plt.legend()
plt.grid(True)
plt.show()

'''Shows yearly sales performance of top BMW models.
The 3 Series remains consistently strong; the X5 shows steady growth.'''
```

```python
# ================================================================
# Seaborn Visualizations
# ================================================================

# Boxplot â€“ Revenue Distribution by Region
sns.boxplot(x="Region", y="Revenue_Million_USD", data=df, palette="Set2")
plt.title("Boxplot - BMW Revenue Distribution by Region")
plt.show()

'''Helps understand revenue variation across regions.
North America has higher median revenue; Europe shows tighter consistency.'''

# Violin Plot â€“ Units Sold by Model
sns.violinplot(x="Model", y="Units_Sold", data=df, palette="muted")
plt.title("Violin Plot - Units Sold by Model")
plt.show()

'''Displays distribution and density of sales per model.
3 Series generally sells more units with less variation than X5.'''
```

```python
# ================================================================
# Interactive Visualizations with Plotly
# ================================================================

# Interactive Scatter Plot â€“ Revenue vs Units Sold
fig = px.scatter(df, x="Units_Sold", y="Revenue_Million_USD",
                 color="Region", size="Revenue_Million_USD",
                 hover_data=["Model", "Year"])
fig.update_layout(title="Interactive Scatter Plot - Revenue vs Units Sold")
fig.show()

'''Visualizes correlation between units sold and revenue.
Hovering shows exact model/year details; larger bubbles = higher revenue.'''

# Interactive Bar Chart â€“ Yearly Sales by Model
fig2 = px.bar(df, x="Year", y="Units_Sold", color="Model", barmode="group",
              text="Units_Sold", title="Interactive Bar Chart - BMW Yearly Sales by Model")
fig2.update_traces(textposition='outside')
fig2.show()

'''Lets you compare year-over-year BMW model performance interactively.
Shows which model dominates in each year.'''
```
